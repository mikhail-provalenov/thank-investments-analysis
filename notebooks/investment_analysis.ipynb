```python
import pandas as pd
import matplotlib.pyplot as plt

print("–ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")
```

```python
# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
df = pd.read_csv("/Users/provalenov_m/Desktop/jupyter_projects/git_jupyter/investment_sample_25k.csv", sep=';', encoding='utf-8')
print(" –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")
print(f"–†–∞–∑–º–µ—Ä —Ç–∞–±–ª–∏—Ü—ã: {df.shape} —Å—Ç—Ä–æ–∫ –∏ —Å—Ç–æ–ª–±—Ü–æ–≤")
```

```python
# –°–º–æ—Ç—Ä–∏–º –Ω–∞ –¥–∞–Ω–Ω—ã–µ
print("–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ —Ç–∞–±–ª–∏—Ü—ã:")
df.head()
```

```python
# –û—Å–Ω–æ–≤–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö
print("–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Å—Ç–æ–ª–±—Ü–∞—Ö:")
df.info()
```

```python
# –ü—Ä–æ—Å—Ç–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
print("–ë–∞–∑–æ–≤–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
df.describe()
```

```python
# –ê–Ω–∞–ª–∏–∑ –≤–æ–∑—Ä–∞—Å—Ç–∞
print("üìà –ê–Ω–∞–ª–∏–∑ –≤–æ–∑—Ä–∞—Å—Ç–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤:")

plt.figure(figsize=(10, 6))
plt.hist(df['age'].dropna(), bins=20, color='lightblue', edgecolor='black')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤')
plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤')
plt.grid(True, alpha=0.3)
plt.show()

print(f"–°—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç: {df['age'].mean():.1f} –ª–µ—Ç")
print(f"–ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç: {df['age'].min()} –ª–µ—Ç")
print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç: {df['age'].max()} –ª–µ—Ç")
```

```python
# –ê–Ω–∞–ª–∏–∑ –¥–æ—Ö–æ–¥–æ–≤
print(" –ê–Ω–∞–ª–∏–∑ –¥–æ—Ö–æ–¥–æ–≤:")


df['monthly_income_amt'] = df['monthly_income_amt'].str.replace(',', '.').astype(float)

plt.figure(figsize=(10, 6))
plt.hist(df['monthly_income_amt'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–µ—Å—è—á–Ω–æ–≥–æ –¥–æ—Ö–æ–¥–∞')
plt.xlabel('–î–æ—Ö–æ–¥ (—Ä—É–±)')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤')
plt.grid(True, alpha=0.3)
plt.show()

print(f"–°—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥: {df['monthly_income_amt'].mean():.0f} —Ä—É–±")
print(f"–ú–µ–¥–∏–∞–Ω–Ω—ã–π –¥–æ—Ö–æ–¥: {df['monthly_income_amt'].median():.0f} —Ä—É–±")
```

```python
# –ê–Ω–∞–ª–∏–∑ –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–æ–Ω–Ω—ã—Ö –ø–æ—Ä—Ç—Ñ–µ–ª–µ–π
print(" –ê–Ω–∞–ª–∏–∑ –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–æ–Ω–Ω—ã—Ö –ø–æ—Ä—Ç—Ñ–µ–ª–µ–π:")


df['portf_total_amt'] = df['portf_total_amt'].str.replace(',', '.').astype(float)

plt.figure(figsize=(10, 6))
plt.hist(df['portf_total_amt'].dropna(), bins=30, color='orange', edgecolor='black')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –ø–æ—Ä—Ç—Ñ–µ–ª—è')
plt.xlabel('–†–∞–∑–º–µ—Ä –ø–æ—Ä—Ç—Ñ–µ–ª—è (—Ä—É–±)')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤')
plt.grid(True, alpha=0.3)
plt.show()

print(f"–°—Ä–µ–¥–Ω–∏–π –ø–æ—Ä—Ç—Ñ–µ–ª—å: {df['portf_total_amt'].mean():.0f} —Ä—É–±")
print(f"–ú–µ–¥–∏–∞–Ω–Ω—ã–π –ø–æ—Ä—Ç—Ñ–µ–ª—å: {df['portf_total_amt'].median():.0f} —Ä—É–±")
```

```python
# –û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–ª–∏–µ–Ω—Ç–æ–≤
print(" –û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–ª–∏–µ–Ω—Ç–æ–≤:")

education_counts = df['education_level_cd'].value_counts()
plt.figure(figsize=(8, 6))
education_counts.plot(kind='bar', color='purple', alpha=0.7)
plt.title('–£—Ä–æ–≤–µ–Ω—å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –∫–ª–∏–µ–Ω—Ç–æ–≤')
plt.xlabel('–£—Ä–æ–≤–µ–Ω—å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()

print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—é:")
print(education_counts)
```

```python
# –°–µ–º–µ–π–Ω–æ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ
print(" –°–µ–º–µ–π–Ω–æ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ –∫–ª–∏–µ–Ω—Ç–æ–≤:")

marital_counts = df['marital_status_cd'].value_counts()
plt.figure(figsize=(8, 6))
marital_counts.plot(kind='bar', color='red', alpha=0.7)
plt.title('–°–µ–º–µ–π–Ω–æ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ –∫–ª–∏–µ–Ω—Ç–æ–≤')
plt.xlabel('–°–µ–º–µ–π–Ω–æ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()

print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Å–µ–º–µ–π–Ω–æ–º—É –ø–æ–ª–æ–∂–µ–Ω–∏—é:")
print(marital_counts)
```

```python
# –ü—Ä–æ—Å—Ç—ã–µ –≤—ã–≤–æ–¥—ã
print(" –ü–†–û–°–¢–´–ï –í–´–í–û–î–´:")
print(f"‚Ä¢ –í—Å–µ–≥–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö: {len(df):,}")
print(f"‚Ä¢ –°—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç: {df['age'].mean():.1f} –ª–µ—Ç")
print(f"‚Ä¢ –°—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥: {df['monthly_income_amt'].mean():,.0f} —Ä—É–±/–º–µ—Å")
print(f"‚Ä¢ –°—Ä–µ–¥–Ω–∏–π –ø–æ—Ä—Ç—Ñ–µ–ª—å: {df['portf_total_amt'].mean():,.0f} —Ä—É–±")
print(f"‚Ä¢ –°–∞–º—ã–π —á–∞—Å—Ç—ã–π —É—Ä–æ–≤–µ–Ω—å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è: {df['education_level_cd'].mode().iloc[0]}")
print(f"‚Ä¢ –°–∞–º—ã–π —á–∞—Å—Ç—ã–π —Å–µ–º–µ–π–Ω—ã–π —Å—Ç–∞—Ç—É—Å: {df['marital_status_cd'].mode().iloc[0]}")

print("\n –ò–¥–µ–∏ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è —Å–µ—Ä–≤–∏—Å–∞:")
print("1. –°–æ–∑–¥–∞—Ç—å –æ–±—Ä–∞–∑–æ–≤–∞—Ç–µ–ª—å–Ω—ã–µ –º–∞—Ç–µ—Ä–∏–∞–ª—ã –¥–ª—è –º–æ–ª–æ–¥—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤")
print("2. –†–∞–∑—Ä–∞–±–æ—Ç–∞—Ç—å —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ –ø—Ä–æ–¥—É–∫—Ç—ã –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã—Ö –≥—Ä—É–ø–ø")
print("3. –£–ø—Ä–æ—Å—Ç–∏—Ç—å onboarding –¥–ª—è –Ω–æ–≤—ã—Ö –∏–Ω–≤–µ—Å—Ç–æ—Ä–æ–≤")
```
